```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import re
```

```python
df = pd.read_csv('data/SFTweets.csv')
df.fillna('null', inplace = True)
```

```python
df.info()
```

```python
#Example Tweet
print(df['text'][0])
```

```python
#Drop duplicate tweets
df.drop_duplicates(subset = 'text', inplace = True)
```

```python
#Cleaning Text (Removing Special Characters, RT, & Redundant Words)
stopwords = {'rt','sanfrancisco','san', 'sf','francisco','california','ca', 'like', 'just', 'people'}

tw_list = pd.DataFrame(df['text'], columns = ['text'])
tw_list['clean'] = tw_list['text']

#Functions to remove redundant words, special characters, and links
wd  = lambda x: ' '.join([word for word in re.split("\W+",x) if word.lower() not in stopwords])
lk = lambda x: re.sub(r'http\S+'," ",x)
rt = lambda x: re.sub("[^A-Za-z0–9 ]"," ",x)
tw_list['clean'] = tw_list['text'].str.lower()
tw_list['clean'] = tw_list['clean'].map(lk).map(rt).map(wd)
df['clean'] = tw_list['clean']
tw_list.head(10)
```

```python
score = []
for tweet in df['clean']:
    score.append(SentimentIntensityAnalyzer().polarity_scores(tweet))

compound = []
neutral = []
negative = []
positive = []

for tweet in score:
    compound.append(tweet['compound'])
    neutral.append(tweet['neu'])
    negative.append(tweet['neg'])
    positive.append(tweet['pos'])

df['compound'] = compound
df['neutral'] = neutral
df['negative'] = negative
df['positive'] = positive
```

```python
vectorizer = CountVectorizer(stop_words = 'english', max_features = 200, ngram_range = (1,2))
vect = vectorizer.fit_transform(df['clean'])
vect_df = pd.DataFrame(vect.toarray(), columns = vectorizer.get_feature_names_out())
vect_df.head()
```

```python
# Most Used Words - I need to play with this a bit to find useful searches
#remove (san francisco, francisco, ca, california, https, etc)
count = pd.DataFrame(vect_df.sum())
countdf = count.sort_values(0,ascending=False)
countdf[1:10]
```
